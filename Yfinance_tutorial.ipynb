- updates are automatic

### What is yfinance?

`yfinance` is a lightweight Python wrapper around Yahoo Finance endpoints.
It allows analysts to retrieve:

- historical price data (daily, weekly, monthly)
- basic company metadata
- financial statements (income statement, balance sheet, cash flow)

It is:
- ✅ free
- ✅ easy to install
- ✅ widely used for prototyping and teaching

⚠️ Important:  
`yfinance` is **not** an institutional-grade data source (e.g. Bloomberg, Refinitiv),
but it is sufficient for learning professional workflows and modelling discipline.

## Retrieving Price Data with yfinance

The core function for historical prices is:

yf.download()


This function requests price data directly from Yahoo Finance servers and returns
a pandas DataFrame.

### Key parameters you must understand

- `tickers`  
  A single ticker (string) or multiple tickers (list)

- `interval`  
  Controls the **frequency** of observations  
  Examples:
  - `"1d"`  → daily data
  - `"1wk"` → weekly data
  - `"1mo"` → monthly data

- `period`  
  Controls **how much history** is returned  
  Examples:
  - `"1y"`   → last year
  - `"5y"`   → last 5 years
  - `"max"`  → full available history

⚠️ Very important rule:
> Never rely on API defaults. Always specify `period` explicitly.

Without this, you may receive only a single month of data, which silently
breaks later analysis.

## Why We Use Monthly Data in This Tutorial

In FINM3422:
- portfolio returns are monthly
- performance metrics are monthly
- Assignment 2 is monthly-frequency

Therefore, we deliberately request:

- `interval = "1mo"` → monthly observations
- `period = "max"`   → long historical window

This ensures:
- correct frequency before computing returns
- sufficient observations for rolling metrics
- consistency with Lecture 6 modelling discipline

Using daily data would require additional resampling
and introduces unnecessary complexity for this tutorial.

## Canonical Price Data Request (Reference)

The following pattern is the **standard, safe way** to retrieve monthly equity prices
using yfinance.

You should understand what each argument does before using it.

```python
%pip install yfinance

import yfinance as yf

tickers = ["AAPL", "MSFT", "GOOGL"]
```

```python
prices = yf.download(
tickers,
period="max",
interval="1mo",
auto_adjust=True,
progress=False
)["Close"]
```

Notes:
- `auto_adjust=True` applies splits and dividends automatically
- `progress=False` disables progress bars (cleaner output)
- selecting `"Close"` returns adjusted closing prices only

The result is a pandas DataFrame where:
- rows = months
- columns = tickers
- index = DatetimeIndex

## Interpreting Missing Values (NA) in API Data

When working with real market data, **missing values are normal**.

Common reasons include:
- companies IPO at different times
- firms did not exist in earlier years
- financial statements are unavailable before listing

Important modelling principle (from Lecture 6):

> NA does **not** mean bad data — it often means the data does not exist economically.

In this tutorial:
- We **keep NA values during ingestion**
- We only remove companies **after defining screening requirements**

This prevents:
- artificial backward filling
- economic nonsense
- biased rankings
